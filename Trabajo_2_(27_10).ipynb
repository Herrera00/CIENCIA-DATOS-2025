```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Coffe_sales.csv")

print(df.shape)
print(df.info())
df.head()
```

```python
print(df.isnull().sum())

print(df['Time_of_Day'].unique())
```

```python
df['Time_of_Day'].value_counts()
df.groupby('Time_of_Day')['money'].describe()

```

```python
# Sumar dinero total por momento del día
ventas_tiempo = df.groupby('Time_of_Day')['money'].sum().sort_values(ascending=False)

# Gráfico de barras
plt.figure(figsize=(7,5))
sns.barplot(x=ventas_tiempo.index, y=ventas_tiempo.values, palette="viridis")
plt.title('Ventas totales según momento del día', fontsize=14)
plt.xlabel('Momento del día')
plt.ylabel('Ingresos totales ($)')
plt.show()

# Gráfico de pastel
plt.figure(figsize=(5,5))
plt.pie(ventas_tiempo, labels=ventas_tiempo.index, autopct='%1.1f%%', colors=sns.color_palette("viridis"))
plt.title('Proporción de ventas por momento del día')
plt.show()
```

#**¿En que momento del dia se vende más cafe?**

**R/:** El analisis indica que las ventas de cafe se concentran principalmente en las horas de la tarde y la noche, mientras que las mañanas presentan una menor demanda.
